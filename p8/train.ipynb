```python
from main import *
from matplotlib import pyplot as plt
from IPython.display import clear_output
```

```python
def plot():
  clear_output()
  plt.plot(range(len(ag.accuracy)), ag.accuracy, linewidth=2.0, color="blue")
  plt.plot(range(len(ag.loss)), ag.loss, linewidth=2.0, color="red")
  plt.plot(range(len(ag.rewards)), ag.rewards, linewidth=2.0, color="yellow")
  plt.show()

get_env(human=False)
ag = Agent(DeepQNet(n_dense=10, dense_sizes=[1024, 512, 512, 256, 128, 64, 32, 16, 16]), on_episode_end=plot)
```

```python
ag.neural_net.net.summary()
```

```python
ag.run_n_steps(int(1e3), human=False)
```

```python
ag.neural_net.net.save_weights("model.weights.h5")
```
